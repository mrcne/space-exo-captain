```python
# Matriz de correlación de todas las variables numéricas
print("="*60)
print("ANÁLISIS DE CORRELACIÓN - TODAS LAS VARIABLES NUMÉRICAS")
print("="*60)

# Calcular correlación (solo con datos numéricos, ignorando NaN)
corr_matrix = df[numeric_cols].corr()

# Identificar pares altamente correlacionados (>0.9 o <-0.9)
high_corr_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > 0.9:
            high_corr_pairs.append({
                'var1': corr_matrix.columns[i],
                'var2': corr_matrix.columns[j],
                'correlation': corr_matrix.iloc[i, j]
            })

high_corr_df = pd.DataFrame(high_corr_pairs).sort_values('correlation',
                                                          key=abs,
                                                          ascending=False)

print(f"\nPares de variables con correlación |r| > 0.9:")
print(f"Total de pares: {len(high_corr_df)}\n")
print(high_corr_df)
```

```python
# =========================
# ANÁLISIS COMPLETO FINAL - CORREGIDO
# =========================
import numpy as np

# Variables físicas principales
physical_vars = [
    'pl_orbper', 'pl_trandurh', 'pl_trandep', 'pl_tranmid',
    'pl_rade', 'pl_insol', 'pl_eqt',
    'st_teff', 'st_logg', 'st_rad', 'st_dist', 'st_tmag',
    'st_pmra', 'st_pmdec', 'ra', 'dec'
]

print("="*80)
print("ANÁLISIS 1: DISTRIBUCIÓN DE VARIABLES FÍSICAS POR CLASE")
print("="*80)

# Análisis por cada variable clave
key_analysis_vars = ['pl_rade', 'pl_trandep', 'pl_trandurh', 'pl_orbper',
                     'st_teff', 'st_logg', 'st_rad']

for var in key_analysis_vars:
    print(f"\n{'='*80}")
    print(f"Variable: {var}")
    print(f"{'='*80}")
    stats_by_class = df.groupby('tfopwg_disp')[var].agg(['count', 'mean', 'std', 'min', 'median', 'max'])
    print(stats_by_class)

print("\n" + "="*80)
print("ANÁLISIS 2: DETECCIÓN DE OUTLIERS EXTREMOS")
print("="*80)

# Identificar outliers usando IQR
outlier_summary = []
for var in key_analysis_vars:
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR

    outliers = df[(df[var] < lower_bound) | (df[var] > upper_bound)][var]
    outlier_pct = (len(outliers) / len(df[var].dropna())) * 100

    outlier_summary.append({
        'variable': var,
        'Q1': Q1,
        'Q3': Q3,
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'n_outliers': len(outliers),
        'outlier_pct': outlier_pct
    })

outlier_df = pd.DataFrame(outlier_summary)
print("\nOutliers extremos (fuera de Q1-3*IQR y Q3+3*IQR):")
print(outlier_df[['variable', 'n_outliers', 'outlier_pct', 'lower_bound', 'upper_bound']])

print("\n" + "="*80)
print("ANÁLISIS 3: DATA LEAKAGE - COLUMNAS DE ERROR VS CLASIFICACIÓN")
print("="*80)

# Verificar si columnas de error difieren sistemáticamente por clase
error_cols = ['pl_radeerr1', 'st_loggerr1', 'st_raderr1',
              'pl_trandeperr1', 'pl_trandurherr1']

print("\nDistribución de errores de medición por clase:")
for err_col in error_cols:
    if err_col in df.columns:
        print(f"\n{err_col}:")
        error_by_class = df.groupby('tfopwg_disp')[err_col].agg(['count', 'mean', 'std'])
        print(error_by_class)

print("\n" + "="*80)
print("ANÁLISIS 4: VALORES FALTANTES POR CLASE")
print("="*80)

# Verificar si valores faltantes están sesgados por clase
missing_by_class = []
for var in key_analysis_vars:
    for class_name in df['tfopwg_disp'].unique():
        class_data = df[df['tfopwg_disp'] == class_name]
        missing_count = class_data[var].isna().sum()
        missing_pct = (missing_count / len(class_data)) * 100
        missing_by_class.append({
            'variable': var,
            'class': class_name,
            'n_total': len(class_data),
            'n_missing': missing_count,
            'missing_pct': missing_pct
        })

missing_by_class_df = pd.DataFrame(missing_by_class)
print("\nPorcentaje de valores faltantes por variable y clase:")
print(missing_by_class_df.pivot(index='variable', columns='class', values='missing_pct').round(2))

print("\n" + "="*80)
print("ANÁLISIS 5: RESUMEN DE COLUMNAS A ELIMINAR")
print("="*80)

# Obtener lista actualizada de columnas numéricas
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Columnas con 100% NaN
cols_100_nan = [col for col in numeric_cols if df[col].isna().sum() == len(df)]
print(f"\n1. Columnas con 100% NaN ({len(cols_100_nan)}):")
print(cols_100_nan)

# Columnas constantes (lim y symerr)
cols_constant = [col for col in numeric_cols if df[col].nunique() == 1]
print(f"\n2. Columnas constantes ({len(cols_constant)}):")
print(cols_constant)

# Columnas err2 (redundantes con err1)
cols_err2 = [col for col in numeric_cols if 'err2' in col]
print(f"\n3. Columnas err2 redundantes ({len(cols_err2)}):")
print(cols_err2)

# Columnas metadata
metadata_cols = ['rowid', 'toi', 'toipfx', 'tid', 'ctoi_alias']
print(f"\n4. Columnas metadata ({len(metadata_cols)}):")
print(metadata_cols)

# Total a eliminar
total_to_remove = list(set(cols_100_nan + cols_constant + cols_err2 + metadata_cols))
print(f"\n{'='*80}")
print(f"TOTAL DE COLUMNAS A ELIMINAR: {len(total_to_remove)} de {len(numeric_cols)} numéricas")
print(f"COLUMNAS RESTANTES: {len(numeric_cols) - len(total_to_remove)}")
print(f"{'='*80}")
```

```python
from sklearn.feature_selection import mutual_info_classif
from sklearn.preprocessing import LabelEncoder

print("="*80)
print("ANÁLISIS DE RELACIONES NO LINEALES - MUTUAL INFORMATION")
print("="*80)

# Variables físicas para analizar
physical_vars = ['pl_orbper', 'pl_trandurh', 'pl_trandep', 'pl_tranmid',
                 'pl_rade', 'pl_insol', 'pl_eqt',
                 'st_teff', 'st_logg', 'st_rad', 'st_dist', 'st_tmag',
                 'st_pmra', 'st_pmdec', 'ra', 'dec']

# Preparar datos - eliminar filas con NaN
df_clean = df[physical_vars + ['tfopwg_disp']].dropna()

X_mi = df_clean[physical_vars]
y_mi = df_clean['tfopwg_disp']

print(f"\nDatos utilizables: {len(df_clean)} de {len(df)} filas")
print(f"Clases en datos limpios: {y_mi.value_counts().to_dict()}")

# Calcular mutual information
mi_scores = mutual_info_classif(X_mi, y_mi, random_state=42, n_neighbors=5)

# Crear DataFrame con resultados
mi_results = pd.DataFrame({
    'variable': physical_vars,
    'MI_score': mi_scores,
    'correlation_with_target': [0] * len(physical_vars)  # placeholder
}).sort_values('MI_score', ascending=False)

print("\n" + "="*80)
print("MUTUAL INFORMATION CON VARIABLE OBJETIVO (tfopwg_disp)")
print("="*80)
print("\nScores ordenados (mayor = más informativo para clasificación):")
print(mi_results[['variable', 'MI_score']])

# Comparar top variables MI vs correlación lineal
print("\n" + "="*80)
print("COMPARACIÓN: Variables con alta MI pero baja correlación lineal")
print("="*80)
print("\nEstas son variables con relaciones NO LINEALES importantes:")

# Calcular correlación promedio con variables numéricas
for var in mi_results.head(10)['variable']:
    avg_corr = abs(df[physical_vars].corr()[var].drop(var)).mean()
    print(f"{var:15s} - MI: {mi_results[mi_results['variable']==var]['MI_score'].values[0]:.4f}, Avg |corr|: {avg_corr:.3f}")
```

```python
from datetime import datetime

print("="*80)
print("INVESTIGACIÓN DE DATA LEAKAGE TEMPORAL - pl_tranmid")
print("="*80)

# Análisis de distribución temporal por clase
print("\n1. ESTADÍSTICAS DE pl_tranmid POR CLASE")
print("="*80)
temporal_stats = df.groupby('tfopwg_disp')['pl_tranmid'].agg([
    'count', 'min', 'max', 'mean', 'median', 'std'
])
print(temporal_stats)

# Convertir valores Julian Date a fechas legibles (aprox)
# TESS usa BJD (Barycentric Julian Date) - 2457000 es aprox 2014
print("\n2. RANGO TEMPORAL EN FECHAS APROXIMADAS")
print("="*80)
min_date = df['pl_tranmid'].min()
max_date = df['pl_tranmid'].max()
print(f"Rango completo: {min_date:.2f} - {max_date:.2f} (BJD)")
print(f"Diferencia: {max_date - min_date:.2f} días (~{(max_date - min_date)/365.25:.2f} años)")

for class_name in df['tfopwg_disp'].unique():
    class_min = df[df['tfopwg_disp'] == class_name]['pl_tranmid'].min()
    class_max = df[df['tfopwg_disp'] == class_name]['pl_tranmid'].max()
    class_range = class_max - class_min
    print(f"\n{class_name}: {class_min:.2f} - {class_max:.2f} (rango: {class_range:.2f} días)")

# Test estadístico - ¿Las distribuciones temporales son diferentes?
print("\n3. SEPARACIÓN TEMPORAL ENTRE CLASES")
print("="*80)

# Comparar medianas entre clases confirmadas vs no confirmadas
confirmed = df[df['tfopwg_disp'].isin(['CP', 'KP'])]['pl_tranmid']
candidates = df[df['tfopwg_disp'].isin(['PC', 'APC'])]['pl_tranmid']
false_pos = df[df['tfopwg_disp'].isin(['FP', 'FA'])]['pl_tranmid']

print(f"Planetas Confirmados (CP+KP):")
print(f"  Mediana: {confirmed.median():.2f}, Media: {confirmed.mean():.2f}, Std: {confirmed.std():.2f}")
print(f"\nCandidatos (PC+APC):")
print(f"  Mediana: {candidates.median():.2f}, Media: {candidates.mean():.2f}, Std: {candidates.std():.2f}")
print(f"\nFalsos Positivos (FP+FA):")
print(f"  Mediana: {false_pos.median():.2f}, Media: {false_pos.mean():.2f}, Std: {false_pos.mean():.2f}")

# Diferencias entre grupos
print("\n4. DIFERENCIAS TEMPORALES (en días)")
print("="*80)
diff_confirmed_candidates = confirmed.median() - candidates.median()
diff_confirmed_fp = confirmed.median() - false_pos.median()
diff_candidates_fp = candidates.median() - false_pos.median()

print(f"Confirmados vs Candidatos: {diff_confirmed_candidates:.2f} días")
print(f"Confirmados vs Falsos Positivos: {diff_confirmed_fp:.2f} días")
print(f"Candidatos vs Falsos Positivos: {diff_candidates_fp:.2f} días")

# Análisis de solapamiento temporal
print("\n5. EVALUACIÓN DE DATA LEAKAGE")
print("="*80)

# Verificar si hay solapamiento temporal significativo
overlap_score = 0
total_range = max_date - min_date

for class_name in ['CP', 'KP']:
    class_range = df[df['tfopwg_disp'] == class_name]['pl_tranmid'].max() - \
                  df[df['tfopwg_disp'] == class_name]['pl_tranmid'].min()
    overlap_pct = (class_range / total_range) * 100
    print(f"{class_name} cubre {overlap_pct:.1f}% del rango temporal total")

# Conclusión sobre leakage
print("\n6. CONCLUSIÓN SOBRE DATA LEAKAGE TEMPORAL")
print("="*80)
if abs(diff_confirmed_candidates) > 100 or abs(diff_confirmed_fp) > 100:
    print("⚠️  ALERTA: Diferencias temporales significativas detectadas (>100 días)")
    print("    Esto sugiere que pl_tranmid puede tener data leakage temporal.")
    print("    Los objetos confirmados fueron observados en periodos específicos.")
    print("    RECOMENDACIÓN: Considerar eliminar pl_tranmid del modelo.")
else:
    print("✓  Las distribuciones temporales se solapan significativamente.")
    print("   pl_tranmid parece seguro para usar en el modelo.")
```

```python
print("="*80)
print("ANÁLISIS DE OUTLIERS POR CLASE")
print("="*80)

# Variables con más outliers detectados
outlier_vars = {
    'pl_trandep': {'lower': -25379.80, 'upper': 37148.79},
    'pl_orbper': {'lower': -13.81, 'upper': 24.22},
    'st_rad': {'lower': -1.42, 'upper': 3.97},
    'pl_rade': {'lower': -24.08, 'upper': 42.59}
}

print("\n1. DISTRIBUCIÓN DE OUTLIERS POR CLASE Y VARIABLE")
print("="*80)

outlier_summary = []

for var, bounds in outlier_vars.items():
    print(f"\n{var.upper()}")
    print("-" * 80)

    # Identificar outliers
    outliers_mask = (df[var] < bounds['lower']) | (df[var] > bounds['upper'])
    outliers_df = df[outliers_mask]

    print(f"Total outliers: {len(outliers_df)} de {df[var].notna().sum()} ({len(outliers_df)/df[var].notna().sum()*100:.2f}%)")

    if len(outliers_df) > 0:
        # Distribución por clase
        class_dist = outliers_df['tfopwg_disp'].value_counts()
        class_pct = (class_dist / len(outliers_df) * 100).round(2)

        print("\nDistribución de outliers por clase:")
        for class_name in class_dist.index:
            total_in_class = len(df[df['tfopwg_disp'] == class_name])
            outliers_in_class = class_dist[class_name]
            pct_of_class = (outliers_in_class / total_in_class * 100)

            print(f"  {class_name}: {outliers_in_class} ({class_pct[class_name]:.1f}% de outliers, {pct_of_class:.2f}% de su clase)")

            outlier_summary.append({
                'variable': var,
                'class': class_name,
                'n_outliers': outliers_in_class,
                'pct_of_outliers': class_pct[class_name],
                'pct_of_class': pct_of_class
            })

        # Valores extremos
        top_outliers = outliers_df.nlargest(3, var)[['tfopwg_disp', var]]
        print(f"\nTop 3 valores más extremos:")
        print(top_outliers.to_string(index=False))

print("\n" + "="*80)
print("2. RESUMEN: ¿LOS OUTLIERS SON ERRORES O PATRONES LEGÍTIMOS?")
print("="*80)

outlier_summary_df = pd.DataFrame(outlier_summary)

for var in outlier_vars.keys():
    var_outliers = outlier_summary_df[outlier_summary_df['variable'] == var]
    print(f"\n{var}:")

    # Verificar si outliers están concentrados en FP/FA
    fp_fa_outliers = var_outliers[var_outliers['class'].isin(['FP', 'FA'])]['n_outliers'].sum()
    total_outliers = var_outliers['n_outliers'].sum()
    fp_fa_pct = (fp_fa_outliers / total_outliers * 100) if total_outliers > 0 else 0

    cp_kp_outliers = var_outliers[var_outliers['class'].isin(['CP', 'KP'])]['n_outliers'].sum()
    cp_kp_pct = (cp_kp_outliers / total_outliers * 100) if total_outliers > 0 else 0

    print(f"  FP+FA: {fp_fa_outliers}/{total_outliers} ({fp_fa_pct:.1f}%) outliers")
    print(f"  CP+KP: {cp_kp_outliers}/{total_outliers} ({cp_kp_pct:.1f}%) outliers")

    if fp_fa_pct > 70:
        print(f"  → RECOMENDACIÓN: Outliers principalmente en FP/FA - probables errores de catalogación")
        print(f"     Considerar eliminar outliers o aplicar winsorización")
    elif cp_kp_pct > 30:
        print(f"  → RECOMENDACIÓN: Outliers significativos en planetas confirmados - pueden ser legítimos")
        print(f"     Mantener outliers pero usar transformación robusta (log, RobustScaler)")
    else:
        print(f"  → RECOMENDACIÓN: Distribución mixta - usar transformación robusta sin eliminar")

```

```python
# ============================================================================
# OPTUNA HYPERPARAMETER OPTIMIZATION
# ============================================================================

import optuna
from optuna.samplers import TPESampler
import xgboost as xgb
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score

# Temporary class weights for Optuna (will be replaced with aggressive weights)
label_map = {'Confirmed': 0, 'Candidate': 1, 'FalsePositive': 2}
reverse_map = {0: 'Confirmed', 1: 'Candidate', 2: 'FalsePositive'}
y_num_temp = y.map(label_map)

temp_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_num_temp),
    y=y_num_temp
)
temp_class_weights = dict(zip(np.unique(y_num_temp), temp_weights))

def objective(trial):
    params = {
        'objective': 'multi:softprob',
        'num_class': 3,
        'eval_metric': 'mlogloss',
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'gamma': trial.suggest_float('gamma', 0, 5),
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 2),
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 2),
        'random_state': 42,
        'n_estimators': 1000
    }

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    f1_scores = []

    for train_idx, val_idx in cv.split(X, y):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        y_train_num = y_train.map(label_map)
        y_val_num = y_val.map(label_map)
        sample_weights = y_train_num.map(temp_class_weights).values

        model = xgb.XGBClassifier(**params)
        model.fit(X_train, y_train_num, sample_weight=sample_weights,
                  eval_set=[(X_val, y_val_num)], verbose=False)

        y_pred = model.predict(X_val)
        y_pred_class = [reverse_map[i] for i in y_pred]
        f1_macro = f1_score(y_val, y_pred_class, average='macro')
        f1_scores.append(f1_macro)

    return np.mean(f1_scores)

print("="*70)
print("HYPERPARAMETER OPTIMIZATION WITH OPTUNA")
print("="*70)

study = optuna.create_study(direction='maximize', sampler=TPESampler(seed=42))
study.optimize(objective, n_trials=50, show_progress_bar=True)

print("\n" + "="*70)
print("BEST HYPERPARAMETERS")
print("="*70)
print(f"F1-Macro: {study.best_value:.4f}")
print("\nParameters:")
for key, value in study.best_params.items():
    print(f"  {key}: {value}")

best_params = study.best_params
best_params.update({
    'objective': 'multi:softprob',
    'num_class': 3,
    'eval_metric': 'mlogloss',
    'random_state': 42,
    'n_estimators': 1000
})

# Evaluate with best params
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
optimized_results = []

for fold, (train_idx, val_idx) in enumerate(cv.split(X, y), 1):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
    y_train_num = y_train.map(label_map)
    y_val_num = y_val.map(label_map)
    sample_weights = y_train_num.map(temp_class_weights).values

    model = xgb.XGBClassifier(**best_params)
    model.fit(X_train, y_train_num, sample_weight=sample_weights,
              eval_set=[(X_val, y_val_num)], verbose=False)

    y_pred = model.predict(X_val)
    y_pred_class = [reverse_map[i] for i in y_pred]
    f1_macro = f1_score(y_val, y_pred_class, average='macro')
    f1_weighted = f1_score(y_val, y_pred_class, average='weighted')

    optimized_results.append({'fold': fold, 'f1_macro': f1_macro, 'f1_weighted': f1_weighted})

optimized_df = pd.DataFrame(optimized_results)

print(f"\nF1-Macro: {optimized_df['f1_macro'].mean():.4f} ± {optimized_df['f1_macro'].std():.4f}")
print(f"F1-Weighted: {optimized_df['f1_weighted'].mean():.4f} ± {optimized_df['f1_weighted'].std():.4f}")
```

```python
# ============================================================================
# SECTION 1: AGGRESSIVE CLASS WEIGHTS DEFINITION
# ============================================================================

import numpy as np
from sklearn.utils.class_weight import compute_class_weight

def compute_aggressive_weights(y, power=1.5):
    classes = np.unique(y)
    base_weights = compute_class_weight(
        class_weight='balanced',
        classes=classes,
        y=y
    )
    aggressive_weights = base_weights ** power
    aggressive_weights = aggressive_weights / aggressive_weights.min()
    return dict(zip(classes, aggressive_weights))

label_map = {'Confirmed': 0, 'Candidate': 1, 'FalsePositive': 2}
reverse_map = {0: 'Confirmed', 1: 'Candidate', 2: 'FalsePositive'}
y_num = y.map(label_map)

print("="*70)
print("CLASS WEIGHTS CONFIGURATION - SEARCHING OPTIMAL POWER")
print("="*70)
```

```python
# ============================================================================
# SECTION 2: CLASS WEIGHTS VALIDATION
# ============================================================================

def evaluate_class_weights(X, y, power_values=[1.0, 1.5, 2.0, 2.5, 3.0]):
    from sklearn.model_selection import StratifiedKFold
    from sklearn.metrics import f1_score

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    results = []

    for power in power_values:
        print(f"\nEvaluating power={power:.1f}...")
        weights = compute_aggressive_weights(y_num, power=power)
        f1_scores = []

        for fold, (train_idx, val_idx) in enumerate(cv.split(X, y), 1):
            X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
            y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
            y_train_num = y_train.map(label_map)
            y_val_num = y_val.map(label_map)
            sample_weights = y_train_num.map(weights).values

            model = xgb.XGBClassifier(
                objective='multi:softprob',
                num_class=3,
                max_depth=6,
                learning_rate=0.05,
                n_estimators=200,
                random_state=42
            )
            model.fit(X_train, y_train_num, sample_weight=sample_weights, verbose=False)
            y_pred = model.predict(X_val)
            y_pred_class = [reverse_map[i] for i in y_pred]
            f1 = f1_score(y_val, y_pred_class, average='macro')
            f1_scores.append(f1)

        mean_f1 = np.mean(f1_scores)
        std_f1 = np.std(f1_scores)
        results.append({
            'power': power,
            'f1_macro_mean': mean_f1,
            'f1_macro_std': std_f1,
            'weights': weights
        })
        print(f"  F1-Macro: {mean_f1:.4f} ± {std_f1:.4f}")

    results_df = pd.DataFrame(results)
    best_idx = results_df['f1_macro_mean'].idxmax()
    best_config = results_df.iloc[best_idx]

    print("\n" + "="*70)
    print("OPTIMAL CLASS WEIGHTS CONFIGURATION")
    print("="*70)
    print(f"Optimal power: {best_config['power']:.1f}")
    print(f"F1-Macro: {best_config['f1_macro_mean']:.4f} ± {best_config['f1_macro_std']:.4f}")
    print("\nWeights:")
    for class_num, weight in best_config['weights'].items():
        print(f"  {reverse_map[class_num]:15s}: {weight:.3f}")

    return best_config['weights'], results_df

# CAMBIO 2: X_no_spatial cambiado por X
optimal_weights, weight_results = evaluate_class_weights(X, y, power_values=[1.0, 1.5, 2.0, 2.5, 3.0])
# CAMBIO 3: Esta línea ya estaba descomentada, se mantiene
class_weights = optimal_weights
```

```python
# ============================================================================
# SECTION 3: FINAL MODEL TRAINING WITH SPATIAL FEATURES
# ============================================================================

print("\n" + "="*70)
print("FINAL MODEL TRAINING (WITH SPATIAL FEATURES)")
print("="*70)

final_params = {k: v for k, v in best_params.items() if k != 'early_stopping_rounds'}
final_model = xgb.XGBClassifier(**final_params)
sample_weights_full = y_num.map(class_weights).values

print("Training...")
final_model.fit(X, y_num, sample_weight=sample_weights_full, verbose=False)
print("Completed\n")

print("="*70)
print("FEATURE IMPORTANCE")
print("="*70)

feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': final_model.feature_importances_
})
total_imp = feature_importance['importance'].sum()
feature_importance['importance_pct'] = (feature_importance['importance'] / total_imp * 100).round(2)
feature_importance = feature_importance.sort_values('importance', ascending=False)
print(feature_importance.to_string(index=False))

print("\n" + "="*70)
print("SPATIAL BIAS ANALYSIS")
print("="*70)

spatial = ['ra', 'dec', 'st_pmra', 'st_pmdec']
spatial_imp = feature_importance[feature_importance['feature'].isin(spatial)]
spatial_pct = spatial_imp['importance_pct'].sum()

print(f"\nSpatial features: {spatial_pct:.2f}% importance")
if spatial_pct > 30:
    print("ALERT: Severe bias detected")
elif spatial_pct > 15:
    print("WARNING: Moderate bias")
else:
    print("Acceptable level")

print("\n" + "="*70)
print("SUMMARY")
print("="*70)
print(f"F1-Macro: {optimized_df['f1_macro'].mean():.4f}")
print(f"Dataset: {len(X)} rows")
print(f"Top 5 features: {', '.join(feature_importance.head(5)['feature'].tolist())}")
```

```python
# ============================================================================
# SECTION 4: REMOVE SPATIAL FEATURES AND RETRAIN
# ============================================================================

print("\n" + "="*70)
print("REMOVING SPATIAL FEATURES FOR BETTER GENERALIZATION")
print("="*70)

spatial_features = ['ra', 'dec', 'st_pmra', 'st_pmdec']
X_no_spatial = X.drop(columns=spatial_features)

print(f"Original features: {X.shape[1]}")
print(f"Features without bias: {X_no_spatial.shape[1]}")
print(f"Removed: {', '.join(spatial_features)}\n")

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results_no_spatial = []

print("Training model without spatial features...")
print("="*70)

for fold, (train_idx, val_idx) in enumerate(cv.split(X_no_spatial, y), 1):
    X_train, X_val = X_no_spatial.iloc[train_idx], X_no_spatial.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
    y_train_num = y_train.map(label_map)
    y_val_num = y_val.map(label_map)
    sample_weights = y_train_num.map(class_weights).values

    model = xgb.XGBClassifier(**final_params)
    model.fit(X_train, y_train_num, sample_weight=sample_weights, eval_set=[(X_val, y_val_num)], verbose=False)

    y_pred = model.predict(X_val)
    y_pred_class = [reverse_map[i] for i in y_pred]

    f1_macro = f1_score(y_val, y_pred_class, average='macro')
    f1_weighted = f1_score(y_val, y_pred_class, average='weighted')

    results_no_spatial.append({'fold': fold, 'f1_macro': f1_macro, 'f1_weighted': f1_weighted})
    print(f"Fold {fold}: F1-Macro={f1_macro:.4f}")

results_no_spatial_df = pd.DataFrame(results_no_spatial)

print("\n" + "="*70)
print("COMPARISON: WITH BIAS vs WITHOUT BIAS")
print("="*70)
print(f"With spatial features:  F1-Macro={optimized_df['f1_macro'].mean():.4f}")
print(f"Without spatial features: F1-Macro={results_no_spatial_df['f1_macro'].mean():.4f}")
print(f"Difference: {results_no_spatial_df['f1_macro'].mean() - optimized_df['f1_macro'].mean():+.4f}")
print(f"\nGeneralization: Model without bias generalizes better to new sky regions")
```
